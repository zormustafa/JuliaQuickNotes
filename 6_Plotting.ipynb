```julia
using Plots
```

```julia
plot(sin, 0, 2pi)
```

```julia
plot([sin, cos], 0, 2pi,  
    label=["sin" "cos"], 
    xlabel="x_values", 
    ylabel="y_values", 
    title="Sin(x) & Cos(x)") 
```

```julia
plot(x -> sin.(x), 0:pi/100:2pi)
```

```julia
x=1:0.1:10;
y=sin.(x);
```

```julia
plot(x, y, color="green", label="sine curve", linewidth=2)   # or lw=2
xlabel!("x values")
ylabel!("y values")
title!("Sine Curve Plotting")

# plot(x, y, color="green", label="sine curve", xlabel="x values", ylabel="y values", title="Sine Curve Plotting")
```

```julia
scatter!(x, y, label="points")    # ! : adding on the pre-existing plot
```

```julia
xflip!()      # flips the x-axis
```

```julia
plot(x,y,seriestype=:scatter,title="My Scatter Plot")
# scatter(x,y,title="My Scatter Plot")
```

```julia
plot(x, x, label="x")
plot!(x, x.^2, label="x_square")
plot!(x, x.^3, label="x_cube")
plot!(x, x.^4, label="x_quad")

# plot(x, [x, x.^2, x.^3, x.^4] , label=["x" "x_square" "x_cube" "x_quad"])
```

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

```julia
b=rand(10,5);
plot(b,linewidth=2,title="My Plot")   # plots matrix, each column as a line
```

```julia
Plots.savefig("myplot.png")         # saves the CURRENT_PLOT in png format
Plots.savefig(p4,"myplot.pdf")      # saves plot from p4 in pdf format
```
