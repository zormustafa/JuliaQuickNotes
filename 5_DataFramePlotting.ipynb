```julia
using DataFrames
using CSV
df = CSV.read("func_data.csv"; header=true); 
```

```julia
first(df,5)
```

```julia
using StatsPlots         # Required for the DataFrame user recipe
# Plot the DataFrame by declaring the points by the column names
@df df plot(:x, [:sin :cos], colour=[:red :yellow])
```

```julia
@df df scatter(:x, [:sin :cos], markersize =2, title="My DataFrame Scatter Plot!")   
```

```julia
@df df scatter(:x, 2 .* :x_square, markersize =2, legend= :bottomright)
```

```julia
@df df plot(:x, cols(5:7), colour = [:red :blue :green], legend= :topleft)
# column can be refered by variable
#  s = :x_root
#  @df df plot(:x, cols(s))

#  cols()  means all columns
```

```julia
df_iris = CSV.read("iris.csv"; header=true);
```

```julia
first(df_iris,5)
```

```julia
@df df_iris marginalhist(:sepallength, :sepalwidth)
```

```julia
@df df_iris corrplot([:sepallength :sepalwidth :petallength :petalwidth], grid = false)

#  @df df_iris corrplot(cols(1:4), grid = false)
```

```julia
cornerplot(Matrix(df_iris[1:4]))   # A correlation plot may also be produced from a matrix
```

```julia
cornerplot(Matrix(df_iris[1:4]), compact=true) 
```

```julia
@df df_iris violin(:class, :sepallength, marker=(0.2,:blue,stroke(0)), legend= :topleft)
@df df_iris boxplot!(:class, :sepallength, marker=(0.3,:orange,stroke(2)))
```

```julia
@df df_iris andrewsplot(:class, cols(1:4))
```
